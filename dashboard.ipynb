```python
from ipyleaflet import Map, GeoJSON, basemaps
import json
import ipywidgets as widgets
from IPython.display import display

# Leer el archivo GeoJSON
with open("regiones_edit.geojson", "r", encoding="utf-8") as f:
    geo_data = json.load(f)

# --------- PESTAÑA 1: Mapa base ---------
mapa_base = Map(
    basemap=basemaps.OpenStreetMap.Mapnik,
    center=(-39.5, -71.5),
    zoom=4,
    scroll_wheel_zoom=False,
    dragging=False,
    zoom_control=False,
    layout=widgets.Layout(width='100%', height='100%')
)

output_map1 = widgets.Box([mapa_base],
    layout=widgets.Layout(
        width='35%',
        height='100%',
        border='2px solid black',
        overflow='visible'
    )
)

# ComboBoxes
combo1a = widgets.Dropdown(
    options=['Opción A1', 'Opción A2', 'Opción A3'],
    description='Combo 1:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='48%')
)

combo2a = widgets.Dropdown(
    options=['Opción B1', 'Opción B2', 'Opción B3'],
    description='Combo 2:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='48%')
)

combos1 = widgets.HBox([combo1a, combo2a],
    layout=widgets.Layout(
        width='100%',
        justify_content='space-between',
        margin='10px 0px'
    )
)

titulo1 = widgets.HTML("<h2 style='text-align:center;'>Dashboard - Pestaña 1</h2>")

caja2a = widgets.HTML(
    value="<div style='background-color:lightgreen; width:100%; height:100%; display:flex; justify-content:center; align-items:center; font-size:16px; font-weight:bold;'>Caja 2</div>",
    layout=widgets.Layout(
        height='49%',
        width='99%',
        border='2px solid blue'
    )
)

caja3a = widgets.HTML(
    value="<div style='background-color:lightblue; width:100%; height:100%; display:flex; justify-content:center; align-items:center; font-size:16px; font-weight:bold;'>Caja 3</div>",
    layout=widgets.Layout(
        height='49%',
        width='99%',
        border='2px solid red'
    )
)

derecha1 = widgets.VBox([caja2a, caja3a],
    layout=widgets.Layout(
        width='63%',
        height='100%',
        justify_content='space-between'
    )
)

contenido1 = widgets.HBox([output_map1, derecha1],
    layout=widgets.Layout(
        width='100%',
        height='650px',
        justify_content='space-between'
    )
)

tab1 = widgets.VBox([titulo1, combos1, contenido1],
    layout=widgets.Layout(
        width='100%',
        padding='10px',
        overflow='hidden',
        height='800px'  # altura aumentada
    )
)

# --------- PESTAÑA 2: Mapa con GeoJSON ---------
mapa_geojson = Map(
    basemap=basemaps.OpenStreetMap.Mapnik,
    center=(-39.5, -71.5),
    zoom=4,
    scroll_wheel_zoom=False,
    dragging=False,
    zoom_control=False,
    layout=widgets.Layout(width='100%', height='100%')
)

geo_layer = GeoJSON(
    data=geo_data,
    style={
        'color': '#2E7D32',
        'weight': 2,
        'fillOpacity': 0.3,
        'fillColor': '#4CAF50'
    },
    hover_style={'fillOpacity': 0.6}
)
mapa_geojson.add_layer(geo_layer)

output_map2 = widgets.Box([mapa_geojson],
    layout=widgets.Layout(
        width='35%',
        height='100%',
        border='2px solid black',
        overflow='visible'
    )
)

# ComboBoxes
combo1b = widgets.Dropdown(
    options=['Opción A1', 'Opción A2', 'Opción A3'],
    description='Combo 1:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='48%')
)

combo2b = widgets.Dropdown(
    options=['Opción B1', 'Opción B2', 'Opción B3'],
    description='Combo 2:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='48%')
)

combos2 = widgets.HBox([combo1b, combo2b],
    layout=widgets.Layout(
        width='100%',
        justify_content='space-between',
        margin='10px 0px'
    )
)

titulo2 = widgets.HTML("<h2 style='text-align:center;'>Dashboard - Pestaña 2</h2>")

caja2b = widgets.HTML(
    value="<div style='background-color:lightgreen; width:100%; height:100%; display:flex; justify-content:center; align-items:center; font-size:16px; font-weight:bold;'>Caja 2</div>",
    layout=widgets.Layout(
        height='49%',
        width='99%',
        border='2px solid blue'
    )
)

caja3b = widgets.HTML(
    value="<div style='background-color:lightblue; width:100%; height:100%; display:flex; justify-content:center; align-items:center; font-size:16px; font-weight:bold;'>Caja 3</div>",
    layout=widgets.Layout(
        height='49%',
        width='99%',
        border='2px solid red'
    )
)

derecha2 = widgets.VBox([caja2b, caja3b],
    layout=widgets.Layout(
        width='63%',
        height='100%',
        justify_content='space-between'
    )
)

contenido2 = widgets.HBox([output_map2, derecha2],
    layout=widgets.Layout(
        width='100%',
        height='650px',
        justify_content='space-between'
    )
)

tab2 = widgets.VBox([titulo2, combos2, contenido2],
    layout=widgets.Layout(
        width='100%',
        padding='10px',
        overflow='hidden',
        height='800px'  # altura igual a tab1
    )
)

# --------- PESTAÑA 3 (placeholder) ---------
tab3 = widgets.HTML("<h3 style='text-align:center; padding:20px;'>Contenido de la pestaña 3</h3>")

# --------- Contenedor de pestañas ---------
tabs = widgets.Tab(
    children=[tab1, tab2, tab3],
    layout=widgets.Layout(width='100%', height='850px')  # altura general mayor
)
tabs.set_title(0, 'Pestaña 1')
tabs.set_title(1, 'Pestaña 2')
tabs.set_title(2, 'Pestaña 3')

display(tabs)
```
