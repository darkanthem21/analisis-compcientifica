```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, HTML, IFrame
import functools
import folium

# Título
titulo = widgets.HTML("<h2 style='text-align:center;'>Dashboard - Pestaña 1</h2>")

# Crear el mapa con Chile centrado y fijo
m = folium.Map(
    location=[-35.6751, -71.5430],  # Centro de Chile
    zoom_start=4, # You might want to adjust this zoom level slightly after seeing the new height
    scrollWheelZoom=False,
    dragging=False,
    zoom_control=False
)

# Cargar el GeoJSON de las regiones de Chile
geojson_path = 'regiones_edit.geojson'
geojson = folium.GeoJson(geojson_path, name="Regiones de Chile").add_to(m)

# Ajustar el mapa al área del GeoJSON (ajusta a todo Chile)
m.fit_bounds(geojson.get_bounds())

# Convertir el mapa a HTML y mostrarlo en la caja
map_html = m._repr_html_()

# --- IMPORTANT CHANGE 1: Make output_map stretch vertically ---
output_map = widgets.Output(
    layout=widgets.Layout(
        width='60%',         # Example width, adjust as needed
        height='auto',        # Allow height to be determined by flexbox
        flex_grow=1           # Allow this widget to grow and fill available vertical space
    )
)

with output_map:
    display(HTML(map_html))

caja1 = output_map

# ComboBox 1 (para Caja 2)
combo1 = widgets.Dropdown(
    options=['Opción A1', 'Opción A2', 'Opción A3'],
    description='Combo 1:',
    style={'description_width': 'initial'}
)

# ComboBox 2 (para Caja 3)
combo2 = widgets.Dropdown(
    options=['Opción B1', 'Opción B2', 'Opción B3'],
    description='Combo 2:',
    style={'description_width': 'initial'}
)

# Layout horizontal de los combobox
combos = widgets.HBox([combo1, combo2], layout=widgets.Layout(justify_content='space-between'))

# Caja 2
caja2 = widgets.HTML(
    value="<div style='background-color:lightgreen; width:100%; height:150px; display:flex; justify-content:center; align-items:center;'>Caja 2</div>",
    layout=widgets.Layout(width='100%', height='150px') # Fixed height for now
)

# Caja 3
caja3 = widgets.HTML(
    value="<div style='background-color:lightblue; width:100%; height:150px; display:flex; justify-content:center; align-items:center;'>Caja 3</div>",
    layout=widgets.Layout(width='100%', height='150px') # Fixed height for now
)

# --- IMPORTANT CHANGE 2: Make the right-side VBox stretch vertically ---
# VBox derecha con cajas 2 y 3
derecha = widgets.VBox(
    [caja2, caja3],
    layout=widgets.Layout(
        width='40%',         # Example width, adjust for balance with map
        flex_grow=1,          # Allow this VBox to grow vertically
        justify_content='flex-start' # Align content to the top
    )
)

# --- IMPORTANT CHANGE 3: Ensure the main content HBox stretches ---
# HBox de contenido (Caja 1 a la izquierda, VBox derecha a la derecha)
contenido = widgets.HBox(
    [caja1, derecha],
    layout=widgets.Layout(
        height='auto',        # Allow height to be determined by flexbox
        flex_grow=1,          # Allow this HBox to grow vertically
        align_items='stretch' # This is crucial: makes children stretch to fill parent's height
    )
)

# --- IMPORTANT CHANGE 4: Ensure the main tab content VBox stretches ---
# VBox total para pestaña 1
tab1 = widgets.VBox(
    [titulo, combos, contenido],
    layout=widgets.Layout(
        height='auto',        # Allow height to be determined by flexbox
        flex_grow=1           # Allow this VBox to grow vertically
    )
)

# Pestañas 2 y 3 (contenidos de ejemplo)
tab2 = widgets.HTML("<h3>Contenido de la pestaña 2</h3>")
tab3 = widgets.HTML("<h3>Contenido de la pestaña 3</h3>")

# --- IMPORTANT CHANGE 5: Set the overall height of the Tab widget ---
# This gives the entire dashboard a defined vertical space to occupy in the notebook.
tabs = widgets.Tab(
    children=[tab1, tab2, tab3],
    layout=widgets.Layout(height='800px') # Adjust this height as needed for your display
)
tabs.set_title(0, 'Pestaña 1')
tabs.set_title(1, 'Pestaña 2')
tabs.set_title(2, 'Pestaña 3')

display(tabs)
```
